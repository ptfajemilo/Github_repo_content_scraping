```python
#Using beautiful soup to get the content in a pretify format
from bs4 import BeautifulSoup
doc = BeautifulSoup(page_contents, 'html.parser')
doc
```

```python
#Extracting the p tag and the a tag. The p tag for the title and description while the a is for the link
selection_class = "f3 lh-condensed mb-0 mt-1 Link--primary"
topic_title_tags = doc.find_all('p',class_ = selection_class )

desc_selector = 'f5 color-fg-muted mb-0 mt-1'
topic_desc_tags = doc.find_all('p', class_=desc_selector)

link_selector  = 'no-underline flex-1 d-flex flex-column'
topic_link_tags = doc.find_all('a', class_= link_selector)
```

```python
#Extracting the topic
topic_titles = []
for topic in topic_title_tags:
    topic_titles.append(topic.text)
```

```python
#Extracting the description of each topic
topic_descriptions = []
for tag in topic_desc_tags:
    topic_descriptions.append(tag.text.strip())
```

```python
#Extracting the links of each topic
topic_urls = []
base_url = 'https://github.com'
for tag in topic_link_tags:
    topic_urls.append(base_url + tag['href'])
```

```python
# Creating a csv file
import pandas as pd
topic_dict ={
    'title': topic_titles,
     'description':topic_descriptions,
     'url': topic_urls
}
```

```python
topic_df = pd.DataFrame(topic_dict)

topic_df.to_csv('topics.csv', index=None)
```

```python
#converting the string to number

def parse_star_count(stars_str):
    stars_str = stars_str.strip()
    if stars_str[-1] == 'k':
        return int(float(stars_str[:-1]) * 1000)
    return int(stars_str)
parse_star_count(star_tags[0].text)
```

```python
#writing a function that can get topic page
def get_topic_page(topic_url):
    #Download the page
    response = requests.get(topic_url)
    #check the response
    if response.status_code != 200:
        raise Exception('Failed to load page {}'.format(topic_url))
    #Parse using the beautiful soup
    topic_doc = BeautifulSoup(response.text, 'html.parser')
    return topic_doc
```

```python
#function that will return the information of repository 
def get_repo_info(repo_tags, star_tags):
    #returns all the required info about a repository
    a_tags = repo_tags.find_all('a')
    username = a_tags[0].text.strip()
    repo_name = a_tags[1].text.strip()
    repo_url = base_url + a_tags[1]['href']
    stars = parse_star_count(star_tags.text)
    return username, repo_name,stars, repo_url 
```

```python
#Writing the function that get the topic repo
def get_topic_repos(topic_doc):
    
    #Get the h3 tag containing the repo title and username
    h3_selection_class = "f3 color-fg-muted text-normal lh-condensed"
    repo_tags = topic_doc.find_all('h3',class_=h3_selection_class)  
    #Get a tag
    a_tags = repo_tags[0].find_all('a')   
    
    #Get the star tag
    star_tags = topic_doc.find_all('span', class_= star_class)
    
    topic_repos_dict = {
    'username': [],
    'repo_name': [],
    'stars': [],
    'repo_url':[]
    }

    #Get repo info
    for i in range(len(repo_tags)):
        repo_info = get_repo_info(repo_tags[i], star_tags[i])
        topic_repos_dict['username'].append(repo_info[0])
        topic_repos_dict['repo_name'].append(repo_info[1])
        topic_repos_dict['stars'].append(repo_info[2])
        topic_repos_dict['repo_url'].append(repo_info[3])
    
    return pd.DataFrame(topic_repos_dict)
```

```python
def scrape_topic(topic_url, path):
    
    if os.path.exists(path):
        print("The file {} already exists. Skipping...".format(path))
        return
    topic_df = get_topic_repos(get_topic_page(topic_url))
    topic_df.to_csv(path + '.csv', index=None)

get_topic_repos(get_topic_page(topic_urls[5])).to_csv('ansible.csv', index=None)
```

```python
#writing a function that can get the topic title
def get_topic_titles(doc):
    selection_class = "f3 lh-condensed mb-0 mt-1 Link--primary"
    topic_title_tags = doc.find_all('p',class_ = selection_class )
    topic_titles = []
    for topic in topic_title_tags:
        topic_titles.append(topic.text)
    return topic_titles
```

```python
#writing a function that will give the description of each topic
def get_topic_desc(doc):  
    desc_selector = 'f5 color-fg-muted mb-0 mt-1'
    topic_desc_tags = doc.find_all('p', class_=desc_selector)
      
    topic_descriptions = []
    for tag in topic_desc_tags:
        topic_descriptions.append(tag.text.strip())
    return topic_descriptions
```

```python
#Writing a function thata will give the url of each topic
def get_topic_urls(doc):
    link_selector  = 'no-underline flex-1 d-flex flex-column'
    topic_link_tags = doc.find_all('a', class_= link_selector)
    topic_urls = []
    base_url = 'https://github.com'
    for tag in topic_link_tags:
        topic_urls.append(base_url + tag['href'])
    return topic_urls
```

```python
def scrape_topics():
    topics_url = 'https://github.com/topics'
    response = requests.get(topics_url)
    if response.status_code !=200:
        raise Exception ('Failed to load page {}'.format(topic_urls))
    topics_dict = {
        'title':get_topic_titles(doc),
        'description': get_topic_desc(doc),
        'url':get_topic_urls(doc)
    }
    
    return pd.DataFrame(topics_dict)
```

```python
def scrape_topics_repos():
    print('Scraping list of topics')
    topics_df = scrape_topics()
    
    os.makedirs('github_repo', exist_ok=True)
    for index, row in topics_df.iterrows():
        print('Scraping top repositories for "{}"'.format(row['title']))
        scrape_topic(row['url'], 'github_repo/{}.csv'.format(row['title']))
        
    
```
